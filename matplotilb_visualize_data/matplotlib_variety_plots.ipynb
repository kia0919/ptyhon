### 바 차트(막대 그래프)
x 축의 값이 분류가 가능한 카테고리 형태일 때 `bar()`, `barh()`메서드로 막대 그래프를 표시할 수 있음  
주의! `bar`메서드는 `x`축의 값이 첫 번째 인수로 옴

```python
y = [2, 3, 1]
x = [1, 2, 3]

plt.bar(x, y)
xlabel = ['가', '나', '다']

plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.show()
```

`xerr`, `yerr`인수를 지정하면 `에러 바(error bar)`를 추가할 수 있음  
`alpha`인수를 지정하면 바의 투명도를 지정할 수 있음(0이 완전 투명, 1이 완전 불투명)

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_position = np.arange(4)
y_performance = 3 + 10 * np.random.rand(4)
error = np.random.rand(4)   # 음수는 들어갈 수 없다, randn은 음수도 들어가서 에러 뜸

plt.title('Bar h')
plt.barh(y_position, y_performance,xerr=error, alpha=0.4)
plt.yticks(y_position, people)
plt.xlabel('x축')
plt.grid(True)  # 그리드로 바 차트 배경에 격자배경 추가.
plt.show()
```

### 파이 차트
각 분류가 전체에서 얼만큼 차지하고 있는지 파악하려 할때 사용하는 차트  
`pie()` 메서드로 표현할 수 있음  


주의!!! 차트가 원형태를 유지할 수 있게 하려면 `axis('equal')`을 지정해야 함

```python
labels = ['개구리', '돼지', '개', '고양이']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0) # 두번째가 돼지 이므로 돼지가 차지한 항목이 파이차트에서 공간이 띄워진다.

plt.title('파이 차트')
# 먼저 파이에 위에서 지정한 전체 사이즈를 정해주고, explode는 위에서 정한 explode로 지정 나머지도 동일, shadow로 그리드 추가,
# autopct는 앞에 몇번째 자리까지 출력할껀지 지정(%i로 바꾸면 정수형), startangle는 첫 데이터 시작 위치 지정 90은  12시 방향 기준
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal') # 지정해야 원형차트가 유지된다.
plt.show()
```

### 스캐터 플롯
두 변수 간의 상관관계를 나타내고자 할 때 사용하는 차트  
`scatter()` 메서드로 표현할 수 있음  
점 하나의 위치는 x, y 값으로 표현됨

```python
# normal(정규분포 함수) 평균 0, 표준편차1, 개수 100개
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.title('스캐터 플롯')
plt.scatter(x, y)
plt.grid(True)
plt.show()
```

만약 x에 대하여 여러 변수를 한 차트로 비교하고자 할 때 점의 크기 혹은 색상으로 추가로 표현할 수 있음  
`scatter()` 메서드의 `c`인수로 색상을, `s`인수로 크기를 지정할 수 있음

```python
n = 30
x = np.random.rand(n)
y1 = np.random.rand(n)
y2 = np.random.rand(n)
y3 = np.pi*(15*np.random.rand(n)) ** 2

plt.title('버블 차트')
# x, y1을 비교대상으로 하고, 컬러값은 y2, 사이즈 값은 y3로 지정
plt.scatter(x, y1, c=y2, s=y3)
plt.grid(True)
plt.show()
```

### ImShow
이미지  데이터 처럼 행과 열을 가진 2차원 데이터에 대하여 각 행렬 값을 색으로 표현하는 차트  
`imshow()`메서드로 표현할 수 있음  

사진을 픽셀 형태로 나타내는 것.

# pip install scikit-learn
from sklearn.datasets import load_digits

```python
digits = load_digits()
X = digits .images[0]
X   # 배열이 출력됨
```

```python
plt.title('mnist digits; 0')
#                                                                cmap은 배경색상을 지정
plt.imshow(X, interpolation='nearest', cmap=plt.cm.bone_r)
# adjust: 사이즈 조절
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65)
plt.xticks([])
plt.yticks([])
plt.show()
```

`cmap`인수로 데이터를 표현하는 칼라맵을 지정할 수 있음  
`interpolation`인수로 데이터의 시각화 표현을 지정할 수 있음 (선명도?)

```python
figure, axes = plt.subplots(1, 4, figsize = (12, 3), subplot_kw={'xticks':[],'yticks':[]}) # 이 문장만 출력해도 공백인 4칸의 이미지가 형성됨
axes[0].set_title('color map Blues')
axes[0].imshow(X, interpolation='nearest', cmap=plt.cm.Blues)

axes[1].set_title('color map Blues_r')
axes[1].imshow(X, interpolation='nearest', cmap=plt.cm.Blues_r)

axes[2].set_title('interpolation none')
axes[2].imshow(X, interpolation='none', cmap=plt.cm.bone_r)

axes[3].set_title('interpolation kaiser')
axes[3].imshow(X, interpolation='kaiser', cmap=plt.cm.bone_r)

plt.show()
```

### 컨투어 플롯
3차원 자료에서 x,y 입력과 z출력이 존재할 때, z 표현을 명암이 아닌 선을 사용하는 방법  
`contour()`, `contourf()`메서드를 사용  
`contour()`: 등고선만 표시 / `contourf()`: 색으로 표시  
입력 x, y에 대하여 1차원 배열을 그대로 사용할 수 없고, `meshgrid()` 메서드 명령으로 포인트 행렬을 만들어 사용해야 한다

```python
def f(x, y):
    return(1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
xx, yy = np.meshgrid(x, y)
zz = f(xx, yy)

plt.title('컨투어 플롯')
plt.contour(xx, yy, zz, colors = 'black') # xx, yy, zz의 선색상을 black 으로 지정
plt.contourf(xx, yy, zz, alpha=0.75, cmap='jet')    # 색상표현방법
plt.grid(True)
plt.show()
```

### 3D 서피스 플롯
3차원 데이터를 입체적으로 표현하는 차트  
3d Axes 객체를 생성하여 사용함  
`plot_wireframe()`, `plot_surface()`메서드로 표현  

```python
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)
xx, yy = np.meshgrid(x, y)
rr = np.sqrt(xx**2 + yy**2)
zz = np.sin(rr)

axes = plt.axes(projection='3d')
axes.set_title('3D 서피스 플롯')
# axes에 플롯 서페이스 명령어에 xx, yy, zz를 넣고, row의 추출간격을 1, column의 sample 추출간격을 1로 지정
axes.plot_surface(xx, yy, zz, rstride=1, cstride=1, cmap='hot')
plt.show()
```
